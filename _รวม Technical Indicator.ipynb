```python
import datetime
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

## Data For Average Calculation

```python
start=datetime.datetime(2010,1,2)
end=datetime.datetime(2014,12,31)
google=web.DataReader('GOOGL','yahoo',start,end)
dfa1_all=google
dfa1_all
```

## MA30

```python
MA30=ta.MA(dfa1_all['Close'], timeperiod=30, matype=0)
MA30=MA30.loc['2011-01-03':'2014-12-31']
MA30=MA30.array
dfa1['MA30']=MA30
```

## MA50

```python
MA50=ta.MA(dfa1_all['Close'], timeperiod=50, matype=0)
MA50=MA50.loc['2011-01-03':'2014-12-31']
MA50=MA50.array
dfa1['MA50']=MA50
```

## MA200

```python
MA200=ta.MA(dfa1_all['Close'], timeperiod=200, matype=0)
MA200=MA200.loc['2011-01-03':'2014-12-31']
MA200=MA200.array
dfa1['MA200']=MA200
```

```python
dfa1
```

## MACD

```python
macd, macdsignal, macdhist = ta.MACD(dfa1_all['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
MACD=macd.loc['2011-01-03':'2014-12-31']
MACD=MACD.array
dfa1['MACD']=MACD
dfa1
```

## RSI

```python
RSI = ta.RSI(dfa1_all['Close'], timeperiod=14)
RSI=RSI.loc['2011-01-03':'2014-12-31']
RSI=RSI.array
dfa1['RSI']=RSI
dfa1
```
