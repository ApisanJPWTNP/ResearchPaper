```python
import talib as ta
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
RSI = ta.RSI(dfa1['Close'])
dfa1['RSI']=RSI

MA30=ta.MA(dfa1['Close'], timeperiod=30)
dfa1['MA30']=MA30

MA50=ta.MA(dfa1['Close'], timeperiod=50, matype=0)
dfa1['MA50']=MA50

MA200=ta.MA(dfa1['Close'], timeperiod=200, matype=0)
dfa1['MA200']=MA200

macd, macdsignal, macdhist = ta.MACD(dfa1['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
MACD=macd.loc['2011-01-03':'2014-12-31']
dfa1['MACD']=MACD
dfa1['macdsignal']=macdsignal
dfa1['macdhist']=macdhist
```

```python
def implement_rsi_strategy(prices, data):    
    buy_price = []
    sell_price = []
    rsi_signal = []
    signal = 0

    for i in range(len(prices)):
        if prices[i] <= 35:
            if signal != 1:
                buy_price.append(prices[i])
                sell_price.append(np.nan)
                signal = 1
                rsi_signal.append(signal)
            else:
                buy_price.append(np.nan)
                sell_price.append(np.nan)
                rsi_signal.append(0)
        elif prices[i] >= 70:
            if signal != -1:
                buy_price.append(np.nan)
                sell_price.append(prices[i])
                signal = -1
                rsi_signal.append(signal)
            else:
                buy_price.append(np.nan)
                sell_price.append(np.nan)
                rsi_signal.append(0)
        else:
            buy_price.append(np.nan)
            sell_price.append(np.nan)
            rsi_signal.append(0)
            
    return buy_price, sell_price, rsi_signal
            
buy_price, sell_price, rsi_signal = implement_rsi_strategy(RSI, dfa1)
```

```python
dfa1["rsi_signal"]=rsi_signal
```
