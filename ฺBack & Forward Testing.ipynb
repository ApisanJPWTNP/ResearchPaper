```python
sl=[]
def make_sl(df):
    for i in range (1,len(df)):
        s_i=df.iloc[i-1]
        sl.append(s_i)
    return (sl)
```

```python
def testing(w,df,sl):
    td=0.3
    mon=100000
    amount=0
    sumall=0
    for i in range(0,len(df)-1):
        decision_d= ((w[0]*sl[i][0]+w[1]*sl[i][1]+w[2]*sl[i][2]+w[3]*sl[i][3]+w[4]*sl[i][4]+w[5]*sl[i][5]+w[6]*sl[i][6]+w[7]*sl[i][7]+w[8]*sl[i][8]+w[9]*sl[i][9])/sum(w))
        if decision_d>td:
            amount=mon//df['Close'][i]
            amount+=sumall
            mon-=sumall*df['Close'][i]
        elif decision_d<-td:
            mon+=amount*df['Close'][i]
            amount=0
        else:
            pass
    return mon
```
