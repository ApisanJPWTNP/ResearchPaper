```python
import talib
```

```python
import yfinance as yf
```

```python
start="2020-01-01"
end="2021-01-01"
data_1=yf.download("SCC.BK")
data_1
```

```python
RSI = talib.RSI(data_1['Close'], timeperiod=14)
RSI
```

```python
RSI.tolist()
```

```python
MA = talib.MA(data_1['Close'], timeperiod=30, matype=0)
MA.tolist()
```

```python
macd, macdsignal, macdhist = talib.MACD(data_1['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
macd.tolist()
```

```python
volume=data_1['Volume']
```

```python
OBV = talib.OBV(data_1['Close'], data_1['Volume'])
```

```python
data = {'RSI': RSI,
        'MA30': MA,
        'MACD': macd,
        'OBV':OBV,
        'Volume':volume}
df=pd.DataFrame(data)
df
```

```python
df.loc[df['RSI'].isna(), ['RSI','MA30','MACD']]
```

```python
df.loc[df['MA30'].isna(), ['RSI','MA30','MACD']]
```
