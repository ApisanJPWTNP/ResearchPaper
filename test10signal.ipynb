```python
count1,count2,count3=0,0,0
decision_d_l=[]
def testing(td,w,df,sl):
    global count1,count2,count3
    mon=100000
    amount=0
    sumall=0
    for i in range(0,len(df)-1):
        decision_d= ((w[0]*sl[i][0]+w[1]*sl[i][1]+w[2]*sl[i][2]+w[3]*sl[i][3]+w[4]*sl[i][4]+w[5]*sl[i][5]+w[6]*sl[i][6]+w[7]*sl[i][7]+w[8]*sl[i][8]
                      +w[9]*sl[i][15])/sum(w))
        # if i!=0:
        #     if df['adj_close_price'][i]<df['adj_close_price'][i-1]*0.2:
        #             mon+=amount*dfo['Close'][i]*0.998
        #             amount=0
        if decision_d>td:
            sumall=mon//df['adj_close_price'][i]
            amount+=sumall
            mon-=sumall*df['adj_close_price'][i]*1.002
            count1+=1
        elif decision_d<-td:
            mon+=amount*df['adj_close_price'][i]*0.998
            amount=0
            count2+=1
        else:
            count3+=1
        decision_d_l.append(mon+amount*df['adj_close_price'][i])
    ret=mon+amount*df['adj_close_price'][len(df)-1]
    return (ret-100000)/1000
```

```python
def make_sl(df):
    sl=[]
    for i in range (1,len(df)):
        s_i=df.iloc[i-1]
        sl.append(s_i)
    return (sl)
```
